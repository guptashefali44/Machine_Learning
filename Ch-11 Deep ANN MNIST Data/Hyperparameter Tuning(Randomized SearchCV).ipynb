```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV
```

```python
from keras.datasets import fashion_mnist
(x_train_full, y_train_full), (x_test, y_test) = fashion_mnist.load_data()

x_val=x_train_full[:5000]/255
x_train=x_train_full[5000:]/255
y_val=y_train_full[:5000]
y_train=y_train_full[5000:]
x_test=x_test/255
```

```python
model=keras.models.Sequential()
```

```python
model.add(keras.layers.Flatten(input_shape=[28,28]))
```

```python
def build_model(n_hidden,n_neurons,learning_rate,input_shape=[28,28]):
    model=keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=input_shape))
    for i in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons,activation="elu"))
    model.add(keras.layers.Dense(10,activation="softmax"))
    optimizer=keras.optimizers.SGD(lr=learning_rate)
    model.compile(loss="sparse_categorical_crossentropy",optimizer=optimizer,metrics=["accuracy"])
    return model


        
    
```

```python
def build_model2(n_hidden,n_neurons,learning_rate,input_shape=[28,28]):
    model=keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=input_shape))
    neuron_list=[]
    neuron_list.append(n_neurons)
    for i in range(1,n_hidden):
        n_new=neuron_list[-1]
        n_neurons=np.random.randint(10,n_new)
        neuron_list.append(n_neurons)
    for neuron in neuron_list:
        model.add(keras.layers.Dense(neuron,activation="elu"))   
    model.add(keras.layers.Dense(10,activation="softmax"))
    optimizer=keras.optimizers.SGD(lr=learning_rate)
    model.compile(loss="sparse_categorical_crossentropy",optimizer=optimizer,metrics=["accuracy"])
    return model
```

```python
keras_class=keras.wrappers.scikit_learn.KerasClassifier(build_model2)
```

```python
reciprocal(3e-4,3e-2)
```

```python
param_dist={"n_hidden":[0,1,2,3],
           "n_neurons":np.arange(10,400),
           "learning_rate":reciprocal(3e-4,3e-2)}

rnd_search_cv=RandomizedSearchCV(keras_class,param_dist,n_iter=10,cv=3)
rnd_search_cv.fit(x_train,y_train,epochs=20,validation_data=(x_val,y_val),callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```
