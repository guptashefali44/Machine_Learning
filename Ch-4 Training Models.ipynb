```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Generate random linear dataset
X=2 * np.random.rand(100, 1)
y=4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.scatter(X,y)
plt.xlabel('x',size=20)
plt.ylabel('Y',size=20)
plt.title('Randomly generated linear dataset')
plt.show()
```

```python
# Lets find optimum parameters using Normal Equation
print(X.shape)
X
```

```python
y=np.ones((100,1))
```

```python
t=np.c_[np.ones((100,1)),X]
t
```

```python
1.# Closed-form equation to calculate model parameters
# Lets solve the normal equation
X_b = np.c_[np.ones((100, 1)), X]     # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
z=np.array([[0], [2]])
z
```

```python
# Now we can make predictions using theta_best
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new] # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
# plot the models predictions
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
#plt.axis([0, 2, 0, 15])
plt.show()
```

## <font color='red'>1. Ridge Regression</font>
### l2 norm


### 1.a Ridge Regression using a closed form solution(Normal equation)
 

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1, solver="cholesky")
ridge_reg.fit(X, y)
ridge_reg.predict([[1.5]])
```

### 1.b Ridge Regression using Stochaistic Gradient Descent

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(penalty="l2")
sgd_reg.fit(X, y.ravel())
sgd_reg.predict([[1.5]])
```

## <font color='red'>2. Lasso Regression(Least absolute Shrinkage and Selection Operator Regression)</font>
### l1 norm


```python
'''Lasso Regression tends to completely eliminate the weights of the least important features (i.e., set them to zero). In
other words, Lasso Regression automatically performs feature selection and outputs a sparse model (i.e., with few nonzero feature weights).'''
```

## 2.a Lasso Regression using Lasso Class


```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)
lasso_reg.predict([[1.5]])
```

# 3. Elastic Net


```python
'''
Elastic Net is a middle ground between Ridge Regression and Lasso Regression.
When r = 0, Elastic Net is equivalent to Ridge Regression, and when r = 1, it is equivalent to Lasso Regression.
'''
```

###  Example using Scikit-Learn’s ElasticNet (l1_ratio corresponds to the mix ratio r):

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X, y)
elastic_net.predict([[1.5]])
```
