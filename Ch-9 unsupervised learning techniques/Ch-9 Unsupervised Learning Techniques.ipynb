### The K-Means Algorithm
* In an unsupervised learning you are given neither the labels nor the centroids,so how can you proceed?
* Just start by placing the centroids randomly (e.g., by picking k instances at random and using their locations as centroids).
* Then label the instances, update the centroids, label the instances, update the centroids, and so on until the centroids stop moving.
* The algorithm is guaranteed to converge in a finite number of steps (usually quite small), it will not oscillate forever2.

**Computational Time Complexity**
* The computational complexity of the algorithm is generally linear with regards to the number of instances m, the number of clusters k and the number of dimensions n. However, this is only true when the data has a clustering structure. If it does not, then in the worst case scenario the complexity can increase exponentially with the number of instances. 

**Centroid Initilisation Methods**
* 1.If you happen to know approximately where the centroids should be (e.g., if you ran another clustering algorithm earlier), then you can set the init hyperparameter to a NumPy array containing the list of centroids, and set n_init to 1.
* 2.Another solution is to run the algorithm multiple times with different random initial‐izations and keep the best solution. This is controlled by the n_init hyperparameter(default value 10).To know which solution is best,it has a model's performance metric called  'model’s inertia'.

**Centroid Initilisation method 1**

### K-Means++

* Take one centroid c(1), chosen uniformly at random from the dataset.
* Choose another centroid such that the instance's distance is inversely proportional to the centroid.This probability distribution ensures that instances further away from already chosen centroids are much more likely be selected as centroids.
* Repeat the previous step until all k centroids have been chosen.



### <font color='red'>Accelerated KMeans and Mini-Batch KMeans</font>

**<font color='red'>Accelearated K-Means</font>**
* <font color='red'>Improvement to the K-Means algorithm was proposed in a 2003 paper by Charles Elkan.</font>
* <font color='red'>It accelerates the algorithm by avoiding many unnecessary distance calculations.</font>
* <font color='red'>This is achieved by exploiting the triangle inequality (i.e., the straight line is always the shortest5) and by keeping track of lower and upper bounds for distances between instances and centroids.</font>

**<font color='red'>Mini-Batch K-Means</font>**
* Another important variant of the K-Means algorithm was proposed in a 2010 paper by David Sculley.
* Instead of using the full dataset at each iteration, the algo‐rithm is capable of using mini-batches.
* This speeds up the algorithm typically by a factor of 3 or 4.
* Scikit-Learn implements this algorithm in the MiniBatchKMeans class.

```python
from sklearn.cluster import MiniBatchKMeans
minibatch_kmeans = MiniBatchKMeans(n_clusters=5)
minibatch_kmeans.fit(X)
```

### Finding the optimal number of clusters
**1. Elbow Curve**
**2. Silhouette_score**

**1. Elbow Curve**



* Inertia is always not a good indicator for choosing optimum number of clusters as inertia keeps on decreasing when the number of clusters increases.
* As we can see in above image :The inertia for k=3 is 653.2, which is much higher than for k=5 (which was 211.6), but with k=8, the inertia is just 119.1. 


* As you can see, the inertia drops very quickly as we increase k up to 4, but then it decreases much more slowly as we keep increasing k.
* There is an elbow at k=4,hence it is the optimum number of clusters.
* This method of choosing optimal number of clusters is more coarse.

**2. Silhouette_score**

* Calculation of Silhouette_score  is more precise and more computationally expensive.
* An instan‐ce’s silhouette coefficient is equal to <font color='red'>(b – a) / max(a, b)</font>
* **a**: the mean distance to the other instances in the same cluster(**inter-cluster distance**).
* **b**:  the mean distance to the instances of the next closest cluster(**intra-cluster distance**) .
* silhouette coefficient can vary from -1 to 1.
* **+1**:The instance is well inside its own cluster and far from other clusters.
* **0**:means that the instance is close to a cluster boundary.
* **-1**: means that the instance may have been assigned to the wrong cluster. 



**Limits of K-Means**

* Despite its merit that K-Means is fast and scalable it is not perfect.
* 1. Algorithm needs to be run several times to avoid sub-optimal solution.
* 2. Finding the optimal number of clusters is a hassle.
* 3. K-Means does not behave very well when the clusters have varying sizes,different densities, or non-spherical shapes.

### Pipeline()

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
 ("kmeans", KMeans(n_clusters=50,random_state=0)),
 ("log_reg", LogisticRegression(random_state=42)),
])
pipeline.fit(X_train, y_train)
```

```python
y_pred=pipeline.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
pipeline.score(X_test, y_test)
```

## What Pipeline does actually

```python
### Training Part
#kmean on X_train
kmeans3=KMeans(n_clusters=50,random_state=0)
kmeans3.fit(X_train)
cluster_labels=kmeans3.predict(X_train)
#use this output as extra feature for logisitic regression along with other features
X_clustered = np.column_stack((X_train, cluster_labels))
print(X_clustered.shape)

#LR on X_train+labels
LR4=LogisticRegression(random_state=42)
LR4.fit(X_clustered, y_train)
```

```python
### Testing Part
cluster_test_labels=kmeans3.predict(X_test)
X_test_clustered = np.column_stack((X_test, cluster_test_labels))
print(X_test_clustered.shape)

LR_test_pred=LR4.predict(X_test_clustered)
print(LR_test_pred.shape, y_test.shape)

accuracy_score(y_test, LR_test_pred)
# LR4.score(LR_test_pred.reshape(-1,1), y_test.reshape(-1,1))
```

```python
# we randomly used n_clusters=50,but we have to find optimum number of clusters.We can do this using gridsearch CV
# Applying gridsearch on pipeline
from sklearn.model_selection import GridSearchCV
param_grid = dict(kmeans__n_clusters=range(2, 100))
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2)
grid_clf.fit(X_train, y_train)
```

**Step 3 label propagation**
* what if we propagated the labels to all the other instances in the same cluster?

```python
y_train_propagated = np.empty(len(X_train), dtype=np.int32)
print(type(y_train_propagated))
print(len(y_train_propagated))
y_train_propagated
```

```python
y_train_propagated[Kmeans4.labels_==3]=y_representative_digits[3]
y_train_propagated[Kmeans4.labels_==3]
```

```python
for i in range(k):
    y_train_propagated[Kmeans4.labels_==i] = y_representative_digits[i]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train_propagated)
log_reg.score(X_test, y_test)
```

```python
percentile_closest = 20
X_cluster_dist = X_digits_dist[np.arange(len(X_train)), Kmeans4.labels_]
for i in range(k):
    in_cluster = (Kmeans4.labels_ == i)
    cluster_dist = X_cluster_dist[in_cluster]
    cutoff_distance = np.percentile(cluster_dist, percentile_closest)
    above_cutoff = (X_cluster_dist > cutoff_distance)
    X_cluster_dist[in_cluster & above_cutoff] = -1

partially_propagated = (X_cluster_dist != -1)
X_train_partially_propagated = X_train[partially_propagated]
y_train_partially_propagated = y_train_propagated[partially_propagated]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train_partially_propagated, y_train_partially_propagated)
log_reg.score(X_test, y_test)
```

### DBScan Clustering
* clustering based on local density estimation.
* This approach allows the algorithm to identify clusters of arbitrary shapes.
* For each instance, the algorithm counts how many instances are located within a small distance  ε (epsilon) from it.
* If an instance has at least min_samples instances in its ε-neighborhood (includ‐ing itself), then it is considered a core instance. In other words, core instances are those that are located in dense regions.
* All instances in the neighborhood of a core instance belong to the same cluster.This may include other core instances, therefore a long sequence of neighboring core instances forms a single cluster.
* Any instance that is not a core instance and does not have one in its neighbor‐hood is considered an anomaly.

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.05)
```
